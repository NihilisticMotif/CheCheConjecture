#SubSpace

##What is Subspace?
* $F=\mathbb{R}\lor F=\mathbb{C}$
* $V=$ Vector Space
* $U\subset V\land U=$ Vector Space $\implies U=$ Subspace of $V$

**Condition of Subspace**
1. $U\subset V$
2. $\vec{0}\in U$
3. $\forall \vec{v},\vec{u}\in U\,:\vec{v}+\vec{u}\in U$
4. $\forall a\in F\land u\in U\,:au\in U$

**Implies that**
* $\therefore U=$ subspace of $V$

**The Smallest and Largest Subspace**
* Size of Subspace = number of all independence variable.
* $\{\vec{0}\}$ is smallest subspace because it contains additive inverse $\vec{0}$ which make it satisfy the properties of vector space.
* $\{\vec{0}\}$ have no independence variable.
* The largest subspace of vector space $V^n$ is the $V^n$ itself.
* $V^n$ have $n$ independence variable.
* $\emptyset$ or empty set is not vector space, because it does not satisfy the property of vector space.




###Prove the condition of Subspace

Given that
* $U\subset V$
* $\forall v,u\in U\,:v+u\in U$

Implies that
* $U$ have Commutativity $\{\vec{v}+\vec{u}=\vec{u}+\vec{v}\}$ and Associativity $\{(\vec{v}+\vec{u})+\vec{w}=\vec{v}+(\vec{u}+\vec{w})\}$ addition properties.

Given that
* $\vec{0}\in U$

Implies that
* $U$ have additive identities $\{\vec{0}\}$

Given that
* $\forall a\in F\land u\in U\,:au\in U$

Implies that
* $U$ have multiplicative identities $\{1\}$
* $U$ have distributive property $\{(a+b)\vec{v}=a\vec{v}+b\vec{v}\land a(\vec{v}+\vec{u})=a\vec{v}+a\vec{u}\}$
* $U$ have scalar multiplicative associativity $\{(ab)\vec{v}=a(b\vec{v})\}$ properties.
* $U$ have additive inverse $\{\vec{v}-\vec{v}=\vec{0}\}$.

Therefore
* $U=\text{VectorSpace}\land U\subset V\implies U=$ SubSpace of $V$

##Theorem

Define
* $V_1,...,V_n=$ Subspace of $V$

Then
* $V_1+...+V_n=$ smallest subspace of $V$ that contains $V_1,...,V_n$



###Proof

Subspace have 4 properties
1. It is the subset of Vector Space
* $U\subset V$
2. Close under Vector Addition
* $\forall \vec{v},\vec{u}\in U\,:\vec{v}+\vec{u}\in U$
3. Close under Scalar Multiplication
* $\forall \vec{u}\in U,\,\forall a\in F\,:a\cdot\vec{u}\in U$
4. Have Additive Identity
* $\vec{0}\in U$

Note that
* $\forall a\in F\,:0\cdot a=0$
* $\forall \vec{v}\in V\,:0\cdot \vec{v}=\vec{0}$

Given that
* $V_1,\,...,\,V_n\in$ Subspace of $V$
* $V=V_1+...+V_n+V_{n+1}+...+V_{n+m}$

Implies that
* $V_1+...+V_n=V_1+...+V_n+0\cdot V_{n+1}+...+0\cdot V_{n+m}$

Therefore
* $V_1+...+V_n$ is the smallest subspace of $V$ that contains $V_1,...,V_n$

* $\text{span}(U_1)+\text{span}(U_2)=\text{span}(U_1\cup U_2)$

Note that Minkowski sum is also defined with in the subset of Vector $A$ and $B$ that might or might not be Subspace of $V$

#Direct Sum

Suppose that
* $V_1,...,V_n=$ SubSpace of $V$

Implies that
* $V_1+...+V_n=$ SubSpace of $V$

If: $\forall \,\vec{v}\in V_1+...+V_n\,:\vec{v}\in\{\vec{v}_1+...+\vec{v}_n|\vec{v}_1\in V_1|\, ...|\, \vec{v}_n\in V_n\}$ have only an unique expression in this form. (where the order does not matter, because every vector space have additive commutative property $\{a+b=b+a\}$)

* $V_1+...+V_n=V_1 \oplus... \oplus V_n$

##Theorem No.1

Suppose that
* $V_1,...,V_n=$ SubSpace of $V$

Implies that
* $V_1\oplus \,...\,\oplus \,V_n\iff [\,\forall \vec{v}_1\in V_1-\{\vec{0}\},\,...,\,\forall\vec{v}_n\in V_n-\{\vec{0}\}\,:\vec{0}\neq\vec{v}_1+...+\vec{v}_n\,]$


###Solution

Given that
* $\forall \vec{v}_i\in V_i-\vec{0},\,\forall \vec{v}_j\in V_j-\vec{0}\,:\vec{0}\neq \vec{v}_i-\vec{v}_j$

Implies that
* $\forall \vec{v}_i\in V_i-\vec{0}\,:\vec{v}_i\not\in V_j$
* $V_i\cap V_j=\{\emptyset,\vec{0}\}$
* $\cap \,V=\{\emptyset,\vec{0}\}$

Suppose that
* $\vec{v}_i\in V_i\,:\vec{a}=\Sigma\,\vec{v}$

Because
* $V_i\cap V_j=\{\emptyset,\vec{0}\}$
* $\cap \,V=\{\emptyset,\vec{0}\}$

Therefore there is only one unique combination of each $\vec{v}_i$ for each $\vec{v}_i\in V_i$ in order to construct $\vec{a}$.
* $V_1\oplus \,...\,\oplus \,V_n\iff [\,\forall \vec{v}_i\in V_i-\{\vec{0}\}\,:\vec{0}\neq\vec{v}_1+...+\vec{v}_n\,]$

On the other hands if
* $\exists V_i,V_j\in$ Set of $V\,:V_i\cap V_j\neq\{\emptyset,\vec{0}\}$

Implies that
* $V_i\cap V_j=\{\emptyset,\vec{0},\vec{v}\}$
* $\exists\vec{v}_i\in V_i\,:\vec{v}_i\in V_j$
* $\vec{0}=\vec{v}_i-1\cdot\vec{v}_j$

Therefore there are several ways to construct $\vec{a}$
* $\forall z\in F\,:\vec{a}=\vec{v}_1+...+z\cdot\vec{v}_i-z\cdot\vec{v}_j+...+\vec{v}_n$
* $\sim[\,V_1\oplus \,...\,\oplus \,V_n\,]\iff [\,\exists V_i,\,\vec{v}_i\in V_i\,:\vec{0}=\vec{v}_1+...+\vec{v}_n\,]$

###Solution

$S$ is Linear Independence
* $\forall \,a_i\subset a\,:\vec{0}=\Sigma\,a\cdot\vec{s}\land a_i=0$

According to this Theorem
* $\forall \vec{v}_i\in S\,:\vec{v}_i\not\in\text{span}(S-\vec{v}_i)\iff S$ is Linear Independence

Implies that
* All of the $\vec{v}\in \text{span}(S)$ can be represented as the linear combination of every $\vec{s}\in S$ uniquely
* $S$ is Linear Independence $\implies\text{span}(S)=\text{span}(\vec{s}_1)\oplus ...\oplus\text{span}(\vec{s}_k)$

On the other hands, all of the $\vec{v}\in \text{span}(S)$ can be represented as the linear combination of every $\vec{s}\in S$ uniquely
* $\forall S\subset V^N\,:\vec{0}\in\text{span}(S)\in\{$ Subspace of $V^N\,\}$
* $\forall a\in F\,:a\cdot0=0$
* $\forall \,a_i\subset a\,:\vec{0}=\Sigma\,a\cdot\vec{s}\land a_i=0$

Therefore
* $S$ is Linear Independence $\iff\text{span}(S)=\text{span}(\vec{s}_1)\oplus ...\oplus\text{span}(\vec{s}_k)$


##Theorem No.4

Prove that
* $\{U\cup W\}$ is Linear Independence $\implies \text{span}(U\cup W)=\text{span}(U)\oplus \text{span}(W)$

###Solution

###Part 1

Define
* $\vec{u}\in U$
* $\vec{w}\in W$
* $\vec{0}\not\in U$
* $\vec{0}\not\in W$
* $U\cap W=\emptyset$

Suppose that
* $U\cup W$ is Linear Independence $\implies U$ is Linear Independence $\land W$ is Linear Independence

Such that
* $\vec{0}=\Sigma\,a_i\cdot \vec{u}_i+\Sigma\,b_i\cdot \vec{w}_i\land a_i=b_i=0$
* $\vec{0}=\Sigma\,a_i\cdot \vec{u}_i\land a_i=0$
* $\vec{0}=\Sigma\,a_i\cdot \vec{w}_i\land a_i=0$

According to this theorem
* $S$ is Linear Independence $\iff\text{span}(S)=\text{span}(\vec{s}_1)\oplus ...\oplus\text{span}(\vec{s}_k)$

Implies that
* $\text{span}(U)=\text{span}(\vec{u}_1)\oplus...\oplus\text{span}(\vec{u}_n)$
* $\text{span}(W)=\text{span}(\vec{w}_1)\oplus...\oplus\text{span}(\vec{w}_m)$

According to this theorem
* $\forall \vec{v}_i\in S\,:\vec{v}_i\not\in\text{span}(S-\vec{v}_i)\iff S$ is Linear Independence

Implies that
* $\forall \vec{u}\in U\,:\vec{u}\not\in \text{span}(U-\vec{u})$
* $\forall \vec{u}\in U\,:\vec{u}\not\in \text{span}(\{U\cup W\}-\vec{u})$
* $\forall \vec{u}\in U\,:\vec{u}\not\in \text{span}(W)$
* $\forall \vec{w}\in W\,:\vec{w}\not\in \text{span}(U)$

Because all linear span is a Vector Space which is also close under Scalar Multiplication.
* $V=$ Vector Sapce
* Subspace $\subset$ Vector Space
* $\forall S\subset V\,:\vec{0}\in \text{span}(S)\in\{$ Subspace of $V\,\}$
* $\forall a\in F\,:\forall \vec{v}\in V\,:a\cdot \vec{v}\in V$

Implies that
* $\forall a\in F\,:\vec{u}\in\text{span}(U)\iff a\cdot\vec{u}\in \text{span}(U)$
* $\forall a\neq 0\,:\vec{u}\not\in\text{span}(W)\iff a\cdot\vec{u}\not\in \text{span}(W)$
* $\text{span}(U)\cap \text{span}(W)=\{\vec{0},\emptyset\}$

Because $\vec{0}$ is the only vector that both $\text{span}(U)$ and $\text{span}(W)$ have, we can conclude that every vector within $\text{span}(U\cup W)$ can be represented uniquely with the linear combination of $U\cup W$ such that
* $\text{span}(U\cup W)=\text{span}(\vec{u}_1)\oplus...\oplus\text{span}(\vec{u}_n)\oplus\text{span}(\vec{w}_1)\oplus...\oplus\text{span}(\vec{w}_m)$
* $U\cup W$ is Linear Independence $\implies\text{span}(U\cup W)=\text{span}(U)\oplus\text{span}(W)$

###Part 2

On the other hands, if
* $\text{span}(U\cup W)=\text{span}(U)\oplus\text{span}(W)$

Note that
* $\vec{0}\in \text{span}(U\cup W)$
* $\vec{0}\in \text{span}(U)$
* $\vec{0}\in \text{span}(W)$

Implies that
* $\text{span}(U)\cap \text{span}(W)=\{\vec{0},\emptyset\}$

Define $B$
* $I=$ Identity Matrix
* $B=\{$ Column $\vec{v}$ of $I\,\}$
* $B$ is Linear Independence

Define $B_i$
* $\vec{b}_i\in\{$ Column $\vec{v}$ of $I\,\}$
* $B_i=\{\vec{b}_i,a\cdot\vec{b}_i\}$
* $B_i$ is Linear Dependence

Because all Linear Span is Vector Space which is close under scalar multiplication, implies that
* $\text{span}(B)=\text{span}(\vec{b}_i)$

According to this theorem
* $\forall \vec{v}_i\in S\,:\vec{v}_i\not\in\text{span}(S-\vec{v}_i)\iff S$ is Linear Independence

Implies that
* $\vec{b}_i\not\in\text{span}(B-\vec{b}_i)$

According to this theorem
* $\forall a\neq 0\,:\vec{u}\not\in\text{span}(W)\iff a\cdot\vec{u}\not\in \text{span}(W)$

Implies that
* $\forall a\neq0\,:\vec{b}_i\not\in\text{span}(B-\vec{b}_i)\iff a\cdot \vec{b}_i\not\in\text{span}(B-\vec{b}_i)$
* $\text{span}(B-\vec{b}_i)\cap \text{span}(B_i)=\{\vec{0},\emptyset\}$

Implies that all of the Vector in $V^N$ can be represented uniquely despite the fact that $B_i$ is Linear Dependence
* $V^N=\text{span}(B-\vec{b}_i)\oplus \text{span}(B_i)$
* $\text{span}(\{B-\vec{b}_i\}\cup B_i)=\text{span}(B-\vec{b}_i)\oplus \text{span}(B_i)$

Therefore
* $\exists U,W\in V^N\,:\text{span}(U\cup W)=\text{span}(U)\oplus\text{span}(W)\land U$ is Linear Dependence




###Therefore

$1^{\text{st}}$ Statement

* $\forall U\subset V^N$
* $\forall \vec{u}\in V^N$
* $\forall a\in F\,:\vec{u}\in\text{span}(U)\iff a\cdot\vec{u}\in \text{span}(U)$
* $\forall a\neq 0\,:\vec{u}\not\in\text{span}(U)\iff a\cdot\vec{u}\not\in \text{span}(U)$

$2^{\text{nd}}$ Statement

* $\forall U,W\subset V^N$
* $U\cup W$ is Linear Independence $\implies\text{span}(U\cup W)=\text{span}(U)\oplus\text{span}(W)$

$3^{\text{rd}}$ Statement

* $\exists U,W\in V^N\,:\text{span}(U\cup W)=\text{span}(U)\oplus\text{span}(W)\land U$ is Linear Dependence

* $\mathbb{R}^n=\{(x_1,...,x_n)\in \mathbb{R}^n|x_1,...,x_n\in\mathbb{R}\}$
* $S^+=\{(x_1,...,x_n)\in \mathbb{R}^n|x_1,...,x_n\in\mathbb{R}^+\}$
* $S^+=\{(x_1,...,x_n)\in \mathbb{R}^n|x_1,...,x_n\in\mathbb{R}^-\}$
* $\forall \vec{v}\in S^+\,:-1\cdot\vec{v}\in S^-$

According to this theorem
* $U,W=$ SubSpace of $V$
* $V=U\oplus W\iff U\cap W=\{\vec{0}\}$

Implies that
* $S^+\cap S^-=\vec{0}\iff \mathbb{R}^n=S^+\oplus S^-$

Define
* $O^+=\{(-x,y)|x,y\in S^+\}$
* $O^-=\{(x,-y)|x,y\in S^+\}$
* $E^+=\{(x,y)|x,y\in S^+\}$
* $E^-=\{(-x,-y)|x,y\in S^+\}$

Given that
* $O^+\cap O^-=\vec{0}$
* $E^+\cap E^-=\vec{0}$

Implies that
* $O=O^+\oplus O^-$
* $E=E^+\oplus E^-$

Given that
* $O,E\in\mathbb{R}^m$
* $O\cap E=\vec{0}$

Therefore
* $\mathbb{R}^m=O\oplus E$

Even if $\mathbb{R}$ is uncountable set, the size of $|\mathbb{R}|$ does not matter, thus we can implies that
* $V_1=\{f_1|\forall x\in\mathbb{R}\,:f_1(-x)=-f_1(x)\}$
* $V_2=\{f_2|\forall x\in\mathbb{R}\,:f_2(-x)=f_2(x)\}$
* $\mathbb{R^\mathbb{R}}=V_1\oplus V_2$
