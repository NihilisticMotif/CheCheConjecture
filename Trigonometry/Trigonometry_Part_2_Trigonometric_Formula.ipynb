##$\cos(\theta)=\cos(\theta+2\pi k)$

###Solution

$2\pi$ is the length of the circumference of the circle with radius $1$.

The Vector $[\,\cos(2\pi k),\sin(2\pi k)\,]$ on the boundary of the unit circle have travel around the unit circle $k$ times.
* $\forall k\in\mathbb{Z}\,:\sin(2\pi k)=\sin(0)=0$
* $\forall k\in\mathbb{Z}\,:\cos(2\pi k)=\cos(0)=1$

According to trigonometry
* $\sin(a+b)=\cos(a)\sin(b)+\sin(a)\cos(b)$

Therefore
* $\sin(\theta+2\pi k)=\cos(\theta)\sin(2\pi k)+\sin(\theta)\cos(2\pi k)$
* $\sin(\theta+2\pi k)=\sin(\theta)$
* $\forall k\in\mathbb{Z}\,:\sin(\theta)=\sin(\theta+2\pi k)$

According to trigonometry
* $\cos(a+b)=\cos(a)\cos(b)-\sin(a)\sin(b)$

Therefore
* $\cos(\theta+2\pi k)=\cos(\theta)\cos(2\pi k)-\sin(\theta)\sin(2\pi k)$
* $\cos(\theta+2\pi k)=\cos(\theta)$
* $\forall k\in\mathbb{Z}\,:\cos(\theta+2\pi k)=\cos(\theta)$

###Therefore

$1^{\text{st}}$ Statement
* $\forall k\in\mathbb{Z}\,:\sin(\theta)=\sin(\theta+2\pi k)$

$2^{\text{nd}}$ Statement
* $\forall k\in\mathbb{Z}\,:\cos(\theta+2\pi k)=\cos(\theta)$


#Even Odd Identity

##$\sin(-\theta)=-\sin(\theta)$

###Solution

###Part 1: If $x,y\in[-\frac{\pi}{2},\frac{\pi}{2}]$

Given that
* $\forall x,y\in[-\frac{\pi}{2},\frac{\pi}{2}]\,:x< y\iff \sin(x) < \sin(y)$

$\begin{matrix}
&\circ&x< y&\iff& \sin(x) < \sin(y)\\
&\circ&-x> -y&\iff& \sin(-x) > \sin(-y)\\
&\circ&x< y&\iff& -\sin(x) > -\sin(y)\\
\end{matrix}$

Suppose that
* $\delta>0$

Implies that
* $x-\delta < x < x+\delta$
* $\sin(x-\delta) < \sin(x) < \sin(x+\delta)$
* $\sin(-x+\delta) > \sin(-x) > \sin(-x-\delta)$
* $-\sin(-x+\delta) < -\sin(-x) < -\sin(-x-\delta)$

Because
* $x-\delta < x < x+\delta\iff \sin(x-\delta) < \sin(x) < \sin(x+\delta)$
* $x-\delta < x < x+\delta\iff-\sin(-x+\delta) < -\sin(-x) < -\sin(-x-\delta)$

We can conclude that
* $\sin(x)=-\sin(-x)$

Therefore
* $-\sin(x)=\sin(-x)$

###Part 2: If $x,y\in[\frac{\pi}{2},\pi]$

Given that
* $\forall x,y\in[\frac{\pi}{2},\frac{3\pi}{2}]\,:x< y\iff \sin(x) > \sin(y)$

Because based on this domain, $\sin(x)$ is also injective function, we can use the same argument to conclude that
* $\sin(x)=-\sin(-x)$

###Part 3: If $x,y\in\mathbb{R}$

According to this theorem
* $\forall k\in\mathbb{Z}\,:\sin(\theta)=\sin(\theta+2\pi k)$

Implies that we can apply this identity in general case where $x\in\mathbb{R}$
* $\forall x\in\mathbb{R}\,:\sin(x)=-\sin(-x)$




##$\cos(-\theta)=\cos(\theta)$

###Solution

Given that
* $\forall k\in\mathbb{Z}\,:\cos(\theta)=\cos(\theta+2\pi k)$
* $\cos(\pm\frac{\pi}{2})=0$
* $\cos(0)=1$
* $\cos(\pi)=-1$

Implies that
* $\forall k\in\mathbb{Z}\,:\cos(2\pi k)=1$
* $\forall k\in\mathbb{Z}\,:\cos(\pi+2\pi k)=-1$

Given that
1. $\cos(x)$ is continuous differentiable function
2. The range of $\cos(x)$ is $[-1,1]$
3. $\forall k\in\mathbb{Z}\,:\pi+2\pi k=-\pi+2\pi(1+k)\land 1+k\in \mathbb{Z}$
4. $\cos(\theta)=\cos(\pi+2\pi k)=\cos(-\pi+2\pi(1+k))=\cos(-\theta)$

Therefore
* $\cos(-\theta)=\cos(\theta)$

##$\tan(-\theta)=-\tan(\theta)$

###Solution

Note that
* $\tan(\theta)=\frac{\sin(\theta)}{\cos(\theta)}$
* $\tan(-\theta)=\frac{\sin(-\theta)}{\cos(-\theta)}$

Given that
* $\sin(-\theta)=-\sin(\theta)$
* $\cos(-\theta)=\cos(\theta)$

Therefore
* $\tan(-\theta)=\frac{-\sin(\theta)}{\cos(\theta)}$
* $\tan(-\theta)=-\tan(\theta)$


#Co-Function Identity

##$\cos(\frac{\pi}{2}-\theta)=\sin(\theta)$

###Solution

Given that
* $\cos(a+b)=\cos(a)\cos(b)-\sin(a)\sin(b)$

$\cos(\frac{\pi}{2}-\theta)=\cos(\frac{\pi}{2})\cos(-\theta)-\sin(\frac{\pi}{2})\sin(-\theta)$

Given that
* $\cos(\pm\frac{\pi}{2})=0$
* $\sin(\frac{\pi}{2})=1$

$\cos(\frac{\pi}{2}-\theta)=-\sin(-\theta)$

Given that
* $\sin(-\theta)=-\sin(\theta)$

$\therefore\cos(\frac{\pi}{2}-\theta)=\sin(\theta)$

##$\sin(\frac{\pi}{2}-\theta)=\cos(\theta)$


###Solution

Given that
* $\sin(a+b)=\sin(b)\cos(a)+\sin(a)\cos(b)$
* $\cos(\pm\frac{\pi}{2})=0$
* $\sin(\frac{\pi}{2})=1$

Implies that
* $\sin(\frac{\pi}{2}-\theta)=\sin(\frac{\pi}{2})\cos(-\theta)+\sin(\theta)\cos(\frac{\pi}{2})$
$\sin(\frac{\pi}{2}-\theta)=\cos(-\theta)$

Given that
* $\cos(-\theta)=\cos(\theta)$

$\therefore\sin(\frac{\pi}{2}-\theta)=\cos(\theta)$


##$\tan(\frac{\pi}{2}-\theta)=\frac{1}{\tan(\theta)}$

###Solution

Given that
* $\tan(\theta)=\frac{\sin(\theta)}{\cos(\theta)}$

Implies that
* $\tan(\frac{\pi}{2}-\theta)=\frac{\sin(\frac{\pi}{2}-\theta)}{\cos(\frac{\pi}{2}-\theta)}$

Given that
* $\cos(\frac{\pi}{2}-\theta)=\sin(\theta)$
* $\sin(\frac{\pi}{2}-\theta)=\cos(\theta)$

Therefore
* $\therefore \tan(\frac{\pi}{2}-\theta)=\frac{\cos(\theta)}{\sin(\theta)}$
* $\therefore \tan(\frac{\pi}{2}-\theta)=\frac{1}{\tan(\theta)}$

#Half Angle Formulas

##$\sin(\frac{\theta}{2})^2$ $=$ $\frac{1-\cos(\theta)}{2}$

###Solution

$\cos(\theta)=\cos(\frac{\theta}{2}+\frac{\theta}{2})$

Given that
* $\cos(a+b)=\cos(a)\cos(b)-\sin(a)\sin(b)$

$\cos(\theta)=\cos(\frac{\theta}{2})\cos(\frac{\theta}{2})-\sin(\frac{\theta}{2})\sin(\frac{\theta}{2})$

$\cos(\theta)=\cos(\frac{\theta}{2})^2-\sin(\frac{\theta}{2})^2$

$-\sin(\frac{\theta}{2})^2=\cos(\theta)-\cos(\frac{\theta}{2})^2$

$\sin(\frac{\theta}{2})^2=-\cos(\theta)+\cos(\frac{\theta}{2})^2$

$\sin(\frac{\theta}{2})^2=-\cos(\frac{\theta}{2})^2+\sin(\frac{\theta}{2})^2+\cos(\frac{\theta}{2})^2$

$\sin(\frac{\theta}{2})^2=-\cos(\frac{\theta}{2})^2+1$

Given that
* $\cos(\theta)=\cos(\frac{\theta}{2})^2-\sin(\frac{\theta}{2})^2$
* $\cos(\theta)+\sin(\frac{\theta}{2})^2=\cos(\frac{\theta}{2})^2$

$\sin(\frac{\theta}{2})^2=-\cos(\theta)-\sin(\frac{\theta}{2})^2+1$

$\sin(\frac{\theta}{2})^2=1-\cos(\theta)-\sin(\frac{\theta}{2})^2$

$2\sin(\frac{\theta}{2})^2=1-\cos(\theta)$

$\therefore \sin(\frac{\theta}{2})^2=\frac{1-\cos(\theta)}{2}$

$\therefore \sin(\frac{\theta}{2})=\pm\sqrt{\frac{1-\cos(\theta)}{2}}$











##$\cos(\frac{\theta}{2})^2$ $=$ $\frac{1+\cos(\theta)}{2}$


###Solution

Given that
* $\cos(\theta)=\cos(\frac{\theta}{2})^2-\sin(\frac{\theta}{2})^2$
* $\cos(\frac{\theta}{2})^2=\cos(\theta)+\sin(\frac{\theta}{2})^2$
* $\sin(\frac{\theta}{2})^2=\frac{1-\cos(\theta)}{2}$

$\cos(\frac{\theta}{2})^2=\cos(\theta)+\frac{1-\cos(\theta)}{2}$

$\cos(\frac{\theta}{2})^2=\frac{1-\cos(\theta)+2\cos(\theta)}{2}$

$\therefore\cos(\frac{\theta}{2})^2=\frac{1+\cos(\theta)}{2}$

$\therefore\cos(\frac{\theta}{2})=\pm\sqrt{\frac{1+\cos(\theta)}{2}}$




##$\tan(\frac{\theta}{2})^2$ $=$ $\frac{1-\cos(\theta)}{1+\cos(\theta)}$


###Solution

Given that
* $\tan(\frac{\theta}{2})^2=\frac{\sin(\frac{\theta}{2})^2}{\cos(\frac{\theta}{2})^2}$
* $\sin(\frac{\theta}{2})^2=\frac{1-\cos(\theta)}{2}$
* $\cos(\frac{\theta}{2})^2=\frac{1+\cos(\theta)}{2}$

$\therefore \tan(\frac{\theta}{2})^2=\frac{1-\cos(\theta)}{1+\cos(\theta)}$

$\therefore \tan(\frac{\theta}{2})=\pm\sqrt{\frac{1-\cos(\theta)}{1+\cos(\theta)}}$


##$\tan(\frac{\theta}{2})$ $=$ $\frac{1-\cos(\theta)}{\sin(\theta)}$


###Solution

Given that
* $\tan(\frac{\theta}{2})=\pm\sqrt{\frac{1-\cos(\theta)}{1+\cos(\theta)}}$

$\tan(\frac{\theta}{2})=\pm\sqrt{\frac{[1-\cos(\theta)]\cdot[1-\cos(\theta)]}{[1+\cos(\theta)]\cdot[1-\cos(\theta)]}}$

$\tan(\frac{\theta}{2})=\pm\sqrt{\frac{[1-\cos(\theta)]^2}{[1+\cos(\theta)]\cdot[1-\cos(\theta)]}}$

$\tan(\frac{\theta}{2})=\pm\sqrt{\frac{[1-\cos(\theta)]^2}{1-\cos(\theta)^2}}$

Given that
* $1=\cos(\theta)^2+\sin(\theta)^2$
* $\sin(\theta)^2=1-\cos(\theta)^2$

$\tan(\frac{\theta}{2})=\pm\sqrt{\frac{[1-\cos(\theta)]^2}{\sin(\theta)^2}}$

$\therefore\tan(\frac{\theta}{2})=\frac{1-\cos(\theta)}{\sin(\theta)}$






##$\tan(\frac{\theta}{2})$ $=$ $\frac{\sin(\theta)}{1+\cos(\theta)}$


###Solution

Given that
* $1=\sin(\theta)^2+\cos(\theta)^2$

$1-\cos(\theta)^2=\sin(\theta)^2$

$1=\frac{1-\cos(\theta)^2}{\sin(\theta)^2}$

$1=\frac{[1-\cos(\theta)]\cdot[1+\cos(\theta)]}{\sin(\theta)^2}$

$\frac{1}{1+\cos(\theta)}=\frac{1-\cos(\theta)}{\sin(\theta)^2}$

$\frac{\sin(\theta)}{1+\cos(\theta)}=\frac{1-\cos(\theta)}{\sin(\theta)}$

Given that
* $\tan(\frac{\theta}{2})=\frac{1-\cos(\theta)}{\sin(\theta)}$

$\therefore \tan(\frac{\theta}{2})=\frac{\sin(\theta)}{1+\cos(\theta)}$
