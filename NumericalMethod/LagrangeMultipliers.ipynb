#**Introduction**

Suppose that:
* $f(\vec{v})\in\mathbb{R}$
* $g(\vec{v})=c\,\land c\in\mathbb{R}$
* $\vec{v}\in\mathbb{R^n}$

We want to maximized or minimized $f(\vec{v})$ by choosing the best $\vec{v}$. We can do this by using this formula

$\frac{\partial f}{\partial \vec{v}}=\lambda \frac{\partial g}{\partial \vec{v}}$

**Note**
* Sometimes, there is no possible $\vec{v}$ that satisfy this condition.

##**Idea**

Suppose that:
* $f(\vec{v})=k,$ we want to maximize $k$
* $g(\vec{v})=c,$ $c$ is the constant

$\mathcal{L}(\vec{v},\lambda)=f(\vec{v})-\lambda(g(\vec{v})-c)$
* We set $\frac{\partial\mathcal{L}(\vec{v},\lambda)}{\partial(\vec{v},\lambda)}=0\implies\mathcal{L}(\vec{v},\lambda)=$ maximum/minimum point
* $\frac{\partial\mathcal{L}(\vec{v},\lambda)}{\partial\lambda}=c-g(\vec{v})=0\implies g(\vec{v})=c$
* $\frac{\partial\mathcal{L}(\vec{v},\lambda)}{\partial\vec{v}}=f'(\vec{v})-\lambda g'(\vec{v})=0\implies f'(\vec{v})=\lambda g'(\vec{v})$

**Note**
* We add $\lambda$ in $\mathcal{L}$, so that we can set the constraint $g(\vec{v})=c$ when optimize $\frac{\partial\mathcal{L}(\vec{v},\lambda)}{\partial(\vec{v},\lambda)}=0$

**Suppose that:**
* $\vec{v}=[x,y]$
* $f(\vec{v})=2x+y$
* $g(\vec{v})=x^2+y^2=1$

We want to find $\vec{v}$ that maximum value of $f(\vec{v})$

We can do it by using this formula: $\frac{\partial f}{\partial \vec{v}}=\lambda \frac{\partial g}{\partial \vec{v}}$

**Find: $x$**

Given that:
* $\frac{\partial f}{\partial x}=2$
* $\frac{\partial g}{\partial x}=2x$
* $\frac{\partial f}{\partial x}=\lambda\frac{\partial f}{\partial x}$

$2=2\lambda x\implies x=\frac{1}{\lambda}$

**Find: $y$**

Given that:
* $\frac{\partial f}{\partial y}=1$
* $\frac{\partial g}{\partial y}=2y$
* $\frac{\partial f}{\partial y}=\lambda\frac{\partial f}{\partial y}$

$1=2\lambda y\implies y=\frac{1}{2\lambda}$

**Find: $\lambda$**

Given that:
* $g(\vec{v})=x^2+y^2=1$

$\therefore \frac{1}{\lambda^2}+\frac{1}{4\lambda^2}=\frac{5}{4\lambda^2}=1$

$\lambda=\frac{\sqrt{5}}{2}$

**Maximized $f(\vec{v})$**

Given that:
* $f(\vec{v})=2x+y$

$f(\vec{v})=\frac{2}{\lambda}+\frac{1}{2\lambda}$

$f(\vec{v})=\frac{5}{2\lambda}$

$f(\vec{v})=\frac{5*2}{2\sqrt{5}}$

$f(\vec{v})=\frac{5}{\sqrt{5}}$

$f(\vec{v})=\frac{\sqrt{5}*\sqrt{5}}{\sqrt{5}}$

$f(\vec{v})=\sqrt{5}$
