#Introduction

###Derivative

**Formula**

$f'(x)=\lim_{h\to0}\frac{f(x+h)-f(x)}{h}$

##Newton Method

**Formula**

$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$
* $n=$ number of iteration
* $f=$ the target function
* $x_n=$ random input
* $x_{n+1}=$ the approximated input value where $f(x_n)≈0$

**Introduction**

The Goal of Newton Method algorithm is to find the input $x_n$ which make the output of function $f(x)$ equal to 0

$f(x_n)=0\,,x_n=?$

**How to find $x_n$?**

1. Choose $x_0$ where $x_0\in\mathbb{R}$
2. Compute $f(x_0)$
3. Create linear function $l(x)$ where $l(x_0)=f(x_0)$

$\therefore l(x)=f'(x_0)x+m=f(x_0)$

$l(x_0)=f'(x_0)x_0+m=f(x_0)$

$m=f(x_0)-f'(x_0)x_0$

$\therefore l(x)=f'(x_0)x+f(x_0)-f'(x_0)x_0$

4. Find input $x_1$ where $l(x_1)=0$

$\therefore0=f'(x_0)x_1+f(x_0)-f'(x_0)x_0$

$0=f'(x_0)x_1-f'(x_0)x_0+f(x_0)$

$0=f'(x_0)(x_1-x_0)+f(x_0)$

$0=x_1-x_0+\frac{f(x_0)}{f'(x_0)}$

$x_0=x_1+\frac{f(x_0)}{f'(x_0)}$

$x_1=x_0-\frac{f(x_0)}{f'(x_0)}$

5. Repeat the Process no.2 to 4

$\therefore x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$
* $n=$ number of iteration
* $f=$ the target function
* $x_n=$ random input
* $x_{n+1}=$ the approximated input value where $f(x_n)≈0$

**Origin of Newton Method**

```python
def f(x):
  return -0.01*x**3+x**2-0.2
def ddxf(x):
  return -0.01*3*x**2+2*x

Xn=1.2
for i in range(10):
  try:
    Xn=Xn-f(Xn)/ddxf(Xn)
  except:
    Xn=Xn
    break
print('x',Xn)
print('f(x)',f(Xn))
```

#Root Finding Algorithm

**Note**

Newton Method can be used for estimating $\sqrt{2},\sqrt{2.5},\sqrt{n}$ where $n=\frac{a}{b}\land a,b\in\mathbb{Z}$

###$\sqrt{2}$

root2.png

https://www.desmos.com/calculator/tcvcx6ri6m?lang=th

**How to find $\sqrt{2}$ using Newton Method?**

1. Let $f(x)=x^2-2$
2. Using Newton Method to find $x_n$ where $f(x)=0$
3. $x_n\approx\sqrt{2}$

```python
def f(x):
  return x**2-2
def ddxf(x):
  return 2*x

Xn=1.2
for i in range(10):
  try:
    Xn=Xn-f(Xn)/ddxf(Xn)
  except:
    Xn=Xn
    break
print('x =',Xn)
print('root 2 =',2**(1/2))
print('f(x) =',f(Xn))
```

###Find Other Root

```python
TargetR=4/9
TargetN=8/9

def f(x):
  return x**(1/TargetR)-TargetN
def ddxf(x):
  return (1/TargetR)*x**((1/TargetR)-1)

Xn=1.2
for i in range(10):
  try:
    Xn=Xn-f(Xn)/ddxf(Xn)
  except:
    Xn=Xn
    break
print('Estimated Number =',Xn)
print('Target Number =',TargetN**(TargetR))
print('f(x) =',f(Xn))
```

#Optimization Algorithm

###Usage

We can use Newton Method to find the lowest point of the function.

```python
start=-5
stop=5
step=2000
X=np.linspace(start, stop, step)

a=0.23
b=-0.31
c=-np.exp(1.5)

# Original Function = Red
plt.plot(X,a*X**2+b*X+c,color='#AA3333')

# First Derivative = Green
plt.plot(X,2*a*X+b,color='#33AA33')

# Second Derivative = Blue
plt.plot(X,2*a*(X/X),color='#3333AA')

plt.grid()
```

**About the Graph**

The red line is the target function $f(x)$, where we want to know the lowest point of the red function. The green line $f'(x)$ is the first ordered derivative of the red function. The blue line $f''(x)$ is the second ordered derivative of the red function.

In order to find the lowest point of the function, we have to find a input $x_n$ where $f'(x_n)=0$

```python
def f(x):
  return 2*a*x+b
def ddxf(x):
  return 2*a

Xn=1.2
for i in range(10):
  try:
    Xn=Xn-f(Xn)/ddxf(Xn)
  except:
    Xn=Xn
    break
print('Estimated Number =',Xn)
print('f(x) =',f(Xn))
```

```python
start=-5
stop=5
step=2000
X=np.linspace(start, stop, step)

a=0.23
b=-0.31
c=-np.exp(1.5)

# Original Function = Red
plt.plot(X,a*X**2+b*X+c,color='#AA3333')

# First Derivative = Green
plt.plot(X,2*a*X+b,color='#33AA33')

# Second Derivative = Blue
plt.plot(X,2*a*(X/X),color='#3333AA')

plt.scatter(Xn,0,color='#AAAA33')
plt.scatter(Xn,a*Xn**2+b*Xn+c,color='#AAAA33')

plt.grid()
```

**Note**

The Limitation of Newton Method is that in order to find the lowest point of $f(x)$, we have to know $f'(x)$ and $f''(x)$. Moreover $f'(x)$ can have multiple $x_n$. So Newton Method does not always guarantee that we will get the lowest point.

###Local Minima Problem

**Note**

Sometimes a function $f(x)$ have several input $x_i$ where $f'(x_i)=0$. So it make the Newtom Method find the local minima instead of global minima.

```python
def f(x):
  return x**3-2*x-1
def ddxf(x):
  return 3*x**2-2

Xn=-np.exp(1)
for i in range(10):
  try:
    Xn=Xn-f(Xn)/ddxf(Xn)
  except:
    Xn=Xn
    break
print('Estimated Number =',Xn)
print('f(x) =',f(Xn))
```

```python
start=-2
stop=2
step=2000
X=np.linspace(start, stop, step)

a=0.23
b=-0.31
c=-np.exp(1.5)

# Original Function = Red
plt.plot(X,(X**4)/4-X**2-X,color='#AA3333')

# First Derivative = Green
plt.plot(X,X**3-2*X-(X/X),color='#33AA33')

# Second Derivative = Blue
plt.plot(X,3*X**2-2*(X/X),color='#3333AA')

plt.scatter(Xn,0,color='#AAAA33')
plt.scatter(Xn,(Xn**4)/4-Xn**2-Xn,color='#AAAA33')  # Local Minimum
plt.axvline(1.618)  # Global Minimum
plt.grid()
```
