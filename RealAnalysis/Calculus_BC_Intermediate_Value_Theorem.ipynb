#Intermediate Value Theorem



Intermediate Value Theorem state that if the function $f(x)$ is continuous over the domain $[a,b]$
* $\exists x\in [a,b]\,:f(a)\leq f(x)\leq f(b)$
* $\exists x\in [a,b]\,:f(x)\in [f(a),f(b)]$

Note that some continuous function $g(x)$ is
* $\exists x\in [a,b]\,:g(x)< f(a)\lor g(x)> f(b)$

* $x$ is continue around $[-1,1]$
* $\exists x\in\mathbb{R}\,:-1\leq x\leq 1$
* $x$ is both surjective and injective function.

According to Intermediate Value Theorem
* $f$ is continuous at $[a,b]$ implies that
* $\exists x\in[a,b]\,:f(a)\leq f(x)\leq f(b)$

Implies that

Therefore there is only a unique of $\cos(x)=x$

* $\exists x\in[a,b]\,:f(a)\leq f(x)\leq f(b)$

##Exercise No.3

Define $f(x)$ such that
* $f(x)$ is continuous on $[0,1]$
* $f(0)=f(1)$
* $\exists n\in\mathbb{N}$

Prove that
* $\exists x\in\mathbb{R}\,:f(x)=f(x+\frac{1}{n})$

###1st Solution

Define $f(x)$ such that
* $f(x)$ is continuous on $[0,1]$
* $f(0)=f(1)$

Because $f(x)$ is continuous over $[0,1]$ implies that
* $\pm\infty\not\in\{f(x)|x\in[0,1]\}$

According to Intermediate Value Theorem
* $f$ is continuous at $[a,b]$ implies that
* $\exists x\in[a,b]\,:f(a)\leq f(x)\leq f(b)$

Implies that
* $\exists x\in[0,1]\,:f(0)= f(x)= f(1)$
* $f(x)$ is not injective.

Therefore
* $\forall n\in\mathbb{N}\,:\exists x\in\mathbb{R}\,:f(x)=f(x+\frac{1}{n})$

###Flaw of this Solution

This solution assume that $f(x)$ is not injective because
1. $f(x)$ is continuous on $[0,1]$
2. $f(0)=f(1)$
3. $\pm\infty\not\in\{f(x)|x\in[0,1]\}$

This solution also assume that
* $\forall n\in\mathbb{N}\,:\exists x\in\mathbb{R}\,:f(x)=f(x+\frac{1}{n})$

The problem is there is no formal proof and theorem that state that

$1^{\text{st}}$ Statement

* $[\,f(a)=f(b)\land f$ is continuous around $[a,b]\,]\implies f(x)$ is not injective.

$2^{\text{nd}}$ Statement

* $\forall n\in \mathbb{N}\,:\forall f\in$ Injective Continuous Function around $[a,b]$
* $\exists x\in[a,b]\,:f(x)=f(x+\frac{1}{n})$




###2nd Solution

Define $f(x)$
* $f(0)=f(1)$
* $f(x)$ is continuous on $[0,1]$

Define $g(x)$
* $\forall n\in\mathbb{N}$
* $g(x)=f(x)-f(x+\frac{1}{n})$
* $S=\{f(0),f(\frac{1}{n}),f(\frac{2}{n}),...,f(\frac{n-1}{n}),f(1)\}$

Note that
* Because of the basic property of limit, $g(x)$ is also continuous on $[0,1]$

Suppose that
* $\forall y\in S\,:y\leq f(\frac{k}{n})$

Implies that
* $g(\frac{k}{n})=f(\frac{k}{n})-f(\frac{k+ 1}{n})\geq0$
* $g(\frac{k-1}{n})=f(\frac{k-1}{n})-f(\frac{k}{n})\leq0$

According to Intermediate Value Theorem
* $f$ is continuous at $[a,b]$ implies that
* $\exists x\in[a,b]\,:f(a)\leq f(x)\leq f(b)$

Implies that
* $0\in[g(\frac{k-1}{n}),g(\frac{k}{n})]$
* $\exists x\in[\frac{k-1}{n},\frac{k}{n}]\,:g(\frac{k-1}{n})\leq g(x)\leq g(\frac{k}{n})$

Because there exists the solution of
* $f(x)-f(x+\frac{1}{n})=0$

Therefore
* $\exists x\in[0,1]\,:\forall n\in\mathbb{N}\,:f(x)=f(x+\frac{1}{n})$
