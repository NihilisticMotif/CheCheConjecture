#Extreme Value

##What is Extreme Value ?

The value of $f(z)$ in interval $[a,b]$ where
* $z\in[a,b]$

Such that
* $\forall x\in[a,b]\,:f(x)\leq f(z)$
* In this case, $z$ is maximum value.

Or Such that
* $\forall x\in[a,b]\,:f(x)\geq f(z)$
* In this case, $z$ is minimum value.

Both maximum and minimum with in the specific domain are both called Extreme Value.

##The Extreme Value Theorem

* $\exists z\in [a,b]\,:f(x)$ is continuous on interval $[a,b]\implies\,f(z)$ is the Minimum or Maximum value of $\{f(x)|x\in[a,b]\}$

##Global Extrema

Function $f$ what have the global maximum or minimum have the following property
1. $f$ is the continuous function or it have the finite number of discontinuity on the interval $[a,b]$ where
* the discontinuious point is not the largest or lowest point $z$ of $f$.
* $f(z)\neq\pm\infty$
* $\lim_{x\to z}f(x)\neq\pm\infty$
2. It is bounded with in the interval $[a,b]$
* $\forall x\in[a,b]\,:f(x)\neq\pm\infty$

Then there exists the unique point $z$
* $z\in[a.b]$

Such that the global maximum of $f$ is
* $\forall x\in[a,b]\,:f(x)\leq f(z)$

Or the global minimum of $f$ is
* $\forall x\in[a,b]\,:f(x)\geq f(z)$

If the function is continuous then we can find Global Extrema by finding all $z$ such that
* $X\in\{x|f'(x)\in[0,\emptyset]\}\cup\{a,b\}$
* $z\in X$
* $f(z)$ is the highest or lowest possible value of $f(x)$ over $[a,b]$

##Local Extrema (a.k.a. Relative Extremum)

The point $x$ is the local extrema if it is the Global Extrema for the arbitrary sufficient small interval $[x-c,x+c]$

Note that Critical Points $x$ of function $f(x)$ is the point where
* $f'(x)=\{0,\emptyset\}$
* $\forall x\in\{$ Critical Points $\}\,:\lim_{z\to x}f(z)=f(x)\neq\{\emptyset\}$

##Applying derivative to find the local extrema.

* $x,z\in[a,b]$
* $\{\,\forall x< z\,:f'(x)>0\,\}\land\{\,\forall x > z\,:f'(x)< 0\,\}\implies z$ is the local maximum.
* $\{\,\forall x< z\,:f'(x)< 0\,\}\land\{\,\forall x > z\,:f'(x)> 0\,\}\implies z$ is the local minimum.

##What is critical point ?

Critical point is the point $x$ where
* $f'(x)\in\{0,\emptyset\}$

Note that not all critical point are extrema value.

One example is
* $f(x)=x^3$
* $f'(x)=3x^2$
* $f'(0)=0$

The critical point of $f(x)$ is $x=0$ but $f(x)$ have no extrema because $f(0)$ is not both largest or least value compare to the point around $x=0$. In addition, $f(x)$ is the continuous injective function that strictly increasing.
