#Definition

##What is Function ?

Function is the map between 2 sets which are domain (input set) and range (output set) such that
* $f\,:X\to Y$
* $\forall x\in X\,:f(x)\in Y$

###Example No.1

Define
* $\vec{v}=(x_1,x_2,x_3,x_4)$

$\vec{v}$ map $\{1,2,3,4\}$ with $F$
* $\vec{v}(1)=x_1$
* $\vec{v}(2)=x_2$
* $\vec{v}(3)=x_3$
* $\vec{v}(4)=x_4$

In general
* $\hat{F}\subset F$
* $\vec{v}\,:\{1,2,3,4\}\to \hat{F}$

###Example No.2

* $f(x)=x^2$

$f(x)$ map $\mathbb{C}$ with $\mathbb{C}$
* $i^2=-1$
* $a,b\in\mathbb{R}$
* $f(a+bi)=a^2-b^2+2abi$

##What is Limit ?

Limit means "approach". The limit of $x$ to $a$ means that the value of $x$ approach to $a$ such that
* $\lim_{x\to a}f(x)=L$

Such that
* $\exists\delta>0$
* $\forall\epsilon>0$
* $|x-a|<\delta\implies |f(x)-L|<\epsilon$

This statement means that the value of $x$ is very close to $a$ such that the different between $x$ and $a$ is nearly $0$
* $\lim_{x\to a}x-a=0$

However, in many case $f(x)$ where $x$ approach $a$ and $f(a)$ give the different output
* $\exists f\in$ Function $\,:\lim_{x\to a}f(x)\neq f(a)$

And in some case the limit of the function does not exists.
* $\exists f\in$ Function$\,:f(x)=\emptyset$
* $\exists f\in$ Function$\,:\lim_{x\to a}f(x)=\emptyset$

Note that if
* $\delta>0\,:f(a+\delta)\not\approx f(a-\delta)\implies \lim_{x\to0}f(a+x)=\emptyset$
* $\lim_{x\to0^+}f(a+x)\neq \lim_{x\to0^-}f(a+x)\implies \lim_{x\to0}f(a+x)=\emptyset$
* $\lim_{x\to a}f(x)=\pm\infty\implies\lim_{x\to a}f(x)=\emptyset$

##Rule of Limit

Here is the rule of limit.

Suppose that
* $f(x)$ and $g(x)$ are both continuous function
* $\lim_{x\to a}f(x)=F$
* $\lim_{x\to a}g(x)=G$
* $c\in \mathbb{R}$

$1^{\text{st}}$ Statement

Sum of 2 Limits
* $\lim_{x\to a}f(x)+g(x)=F+G$

$2^{\text{nd}}$ Statement

Limit multiplied by constant
* $\lim_{x\to a}cf(x)=cF$

$3^{\text{rd}}$ Statement

Product of 2 Limit
* $\lim_{x\to a}f(x)g(x)=FG$

$4^{\text{th}}$ Statement

Limit division
* $\lim_{x\to a}\frac{f(x)}{g(x)}=\frac{F}{G}\iff G\neq0$

$5^{\text{th}}$ Statement

Limit power
* $\forall n\in\mathbb{N}\,:\lim_{x\to a}f(x)^n=F^n$

$6^{\text{th}}$ Statement

Limit Odd Root
* $\forall n\in\{2z+1|z\in\mathbb{Z}\}\,:\lim_{x\to a}f(x)^\frac{1}{n}=F^\frac{1}{n}$

$7^{\text{th}}$ Statement

Limit Even Root
* $\forall n\in\{2z|z\in\mathbb{Z}\}\,:\lim_{x\to a}f(x)^\frac{1}{n}=F^\frac{1}{n}\iff F\geq 0$

Note that because limit have those properties, sets of all the function with the real and complex number are also the vector space that have infinite uncountable dimension. The discipline that study infinite dimensional vector space is called "Functional Analysis".

##What is Indeterminate Form ?

Indeterminate Form is the mathematical expression that cannot compute meaningful result with only direct substitution.

On the other hands, the Determinate Form is the mathematical expression can compute the result as real number, complex number or infinity.

$1^{\text{st}}$ Example of Determinate Form

* $\lim_{x\to a}f(x)=0$
* $\lim_{x\to a}g(x)=0$

Then
* $\lim_{x\to a}f(x)\cdot g(x)=0$
* $0\cdot 0=0$

$2^{\text{nd}}$ Example of Determinate Form

* $\lim_{x\to a}f(x)=\infty$
* $\lim_{x\to a}g(x)=0$

Because
* $\lim_{x\to0^+}\frac{1}{x}=\infty$

Then
* $\lim_{x\to a}\frac{f(x)}{g(x)}=\infty\cdot\infty$
* $\frac{\infty}{0}=\infty$

$1^{\text{st}}$ Example of Indeterminate Form

* $\lim_{x\to a}f(x)=0$
* $\lim_{x\to a}g(x)=0$

Then
* $\lim_{x\to a}\frac{f(x)}{g(x)}=\,?$
* $\frac{0}{0}$

$2^{\text{nd}}$ Example of Indeterminate Form

* $\lim_{x\to a}f(x)=\infty$
* $\lim_{x\to a}g(x)=0$

Then
* $\lim_{x\to a}f(x)\cdot g(x)=\,?$
* $0\cdot\infty$

##Common Technique to Solve Limit

$1^{\text{st}}$ Method: L'Hospital Rule

* $\lim_{x\to a}\frac{f(x)}{g(x)}=\lim_{x\to a}\frac{f'(x)}{g'(x)}$

$2^{\text{nd}}$ Method: Factoring

* If: $f(x)=g(x)k(x)$
* $\lim_{x\to a}\frac{f(x)}{g(x)}=\lim_{x\to a}\frac{g(x)k(x)}{g(x)}=k(x)$

$3^{\text{rd}}$ Method: Substitution

$4^{\text{th}}$ Method: Trigonometric Identity

##Exercise No.5

Find
* $\lim_{x\to1}\frac{(2x)^{\frac{1}{2}}(x-1)}{|x-1|}=\,?$

###Wrong Solution

* $\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{|x-1|}=\,?$

According to L'Hospital Rule
* $\lim_{x\to a}\frac{f(x)}{g(x)}=\lim_{x\to a}\frac{f'(x)}{g'(x)}$

Given that
* $\frac{d}{dx}[\sqrt{x}]=\frac{1}{2\sqrt{x}}=\frac{\sqrt{x}}{2x}$
* $\frac{d}{dx}[\sqrt{2x}]=\frac{\sqrt{2x}}{2x}$
* $\frac{d}{dx}[x-1]=1$
* $\frac{d}{dx}[f(x)g(x)]=f(x)g'(x)+g(x)f'(x)$

Implies that
* $\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{|x-1|}=\lim_{x\to1}\sqrt{2x}+(x-1)(\frac{\sqrt{2x}}{2x})$
* $\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{|x-1|}=\lim_{x\to1}\sqrt{2x}+(\frac{\sqrt{2}}{2}-\frac{\sqrt{2x}}{2x})$
* $\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{|x-1|}=\sqrt{2}$

Therefore
* $\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{|x-1|}=\sqrt{2}$

Reference
* https://www.cuemath.com/calculus/derivative-of-root-x/


###Why This Solution is Wrong?

The right solution is
* $\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{|x-1|}=\emptyset$

The reason is that
* $f(x)=|x|\,:f'(x)=1\iff x\neq0$
* $f(x)=|x|\,:f'(0)=\emptyset$

Thus we cannot directly apply L'Hospital Rule

The another reason is that
* $\lim_{x\to1^+}\frac{\sqrt{2x}(x-1)}{|x-1|}=\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{(x-1)}=\sqrt{2x}=\sqrt{2}$
* $\lim_{x\to1^-}\frac{\sqrt{2x}(x-1)}{|x-1|}=\lim_{x\to1}\frac{\sqrt{2x}(x-1)}{-1\cdot (x-1)}=-\sqrt{2x}=-\sqrt{2}$

#Limit Exercise Part 2: Limit Approach $\infty$

Some function $f(x)$ have limit at infinity if and only if
* $\lim_{x\to\infty}f(x)=L$
* $\forall\epsilon>0\,:\exists x\in(0,\infty)\,:|f(x)-L|<\epsilon$

On the other hands, some function $f(x)$ also have limit at minus infinity such that
* $\lim_{x\to-\infty}f(x)=M$
* $\forall\epsilon>0\,:\exists x\in(-\infty,0)\,:|f(x)-M|<\epsilon$

### What is asymptotes?

An asymptote is a straight line that constantly approaches a given curve but does not meet at any infinite distance.

#Continuous Function

The function $f(x)$ is continuous on point $a$ and interval $[x_0,x_1]$ if it satisfy 4 properties
1. It is defined on $a$
* $f(a)\in\mathbb{R}$
2. Its limit is defined on $a$
* $\lim_{x\to a}f(x)\in\mathbb{R}$
3. $f(a)$ is equal to its limit on $a$
* $\lim_{x\to a}f(x)=f(a)$
4. $\forall x\in [x_0,x_1]$
* $\forall\epsilon>0\,:\exists\delta>0\,:|x-a|<\delta\implies|f(x)-f(a)|<\epsilon$

Reference
* https://brilliant.org/wiki/continuous-functions/

#Uniform Continuity

A function is uniformly continuous if it preserves small distances uniformly across its domain. In other word, suppose that
* Function is continuous in $[a,b]$
* $\forall x,y\in[a,b]$
* $\forall\epsilon>0\,:\exists\delta>0\,:|x-y|<\delta\implies |f(x)-f(y)|<\epsilon$

If $\delta$ is not depending on $\epsilon$ implies that the function is uniformly continuous.

Note that
* $f(x)$ is uniformly continuous over $[a,b]\implies\,f(x)$ is continuous over $[a,b]$

Reference
* https://www.tutorocean.com/questions-answers/what-is-an-uniformly-continuous-function
* https://youtu.be/PA0EJHYymLE?si=7Lbt0KOcLaxWkOTO

###Example No.1

Define
* $f(x)=mx+c$
* $x,y\in[a,b]$

Then
* $|f(x)-f(y)|<\epsilon$
* $|m(x-y)|<\epsilon$
* $|x-y|<\delta=\frac{\epsilon}{m}$

Implies that
* $\delta$ is not depending on the $x$ and $y$
* $f(x)$ is uniform continuous

##Note that

* $\sqrt{x}$ is uniform continuous over $[0,\infty)$
* $\frac{1}{x}$ is uniform continuous over $[2,\infty)$
* $\frac{1}{x}$ is not uniform continuous over $(0,1]$
