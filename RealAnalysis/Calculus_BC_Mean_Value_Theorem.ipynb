#Mean Value Theorem

Suppose that
* $f(x)$ is continuous over the interval $[a,b]$
* $f(x)$ is differentible over the interval $(a,b)$

Then
* $\exists c\in[a,b]\,:f'(c)=\frac{f(a)-f(b)}{a-b}$
* Where $f'(c)=$ the average of change of $\{f'(x)|x\in[a,b]\}$

##Exercise No.1

Determine all the numbers $c$ that satisfy the conclusion of the mean value theorem for this function
* $f(x)=x^3+2x^2+x$ on the interval $[-2,2]$

Reference
* https://brilliant.org/wiki/mean-value-theorem/

###Solution

Calculate $f(-2)$ and $f(2)$
* $f(-2)=-8+8-2=-2$
* $f(2)=8+8+2=18$
* $\frac{f(-2)-f(2)}{-2-2}=\frac{18+2}{2+2}=\frac{20}{4}=5$

Calculate $f'(x)$
* $f'(x)=3x^2+4x+1$
* $3x^2+4x+1=5$
* $3x^2+4x-4=0$
* $(x+2)(3x-2)=0\iff x\in\{-2,\frac{2}{3}\}$

Therefore
* $f'(-2)=12-8+1=5=\frac{f(-2)-f(2)}{-2-2}$
* $f'(\frac{2}{3})=\frac{4}{3}+\frac{8}{3}+\frac{3}{3}=\frac{15}{3}=5=\frac{f(-2)-f(2)}{-2-2}$
* $\{-2,\frac{2}{3}\}\subset[-2,2]$

Find $\frac{d}{dx}[\sqrt{x-1}]$

$\frac{d}{dx}[\sqrt{x-1}]=\lim_{h\to0}\frac{\sqrt{x+h-1}-\sqrt{x-1}}{h}$

$\frac{d}{dx}[\sqrt{x-1}]=\lim_{h\to0}\frac{(\sqrt{x+h-1}-\sqrt{x-1})(\sqrt{x+h-1}+\sqrt{x-1})}{h(\sqrt{x+h-1}+\sqrt{x-1})}$

$\frac{d}{dx}[\sqrt{x-1}]=\lim_{h\to0}\frac{x+h-1-x+1}{h\sqrt{x+h-1}+h\sqrt{x-1}}$

$\frac{d}{dx}[\sqrt{x-1}]=\lim_{h\to0}\frac{h}{h\sqrt{x+h-1}+h\sqrt{x-1}}$

$\frac{d}{dx}[\sqrt{x-1}]=\lim_{h\to0}\frac{1}{\sqrt{x+h-1}+\sqrt{x-1}}$

$\frac{d}{dx}[\sqrt{x-1}]=\frac{1}{2\sqrt{x-1}}$

#Mean Value Theorem for Integrals

Suppose that
* $f(c)$ is the continuous function on the interval $[a,b]$ and differentiable on the interval $(a,b)$
* $F'(x)=f(c)$

According to Mean Value Theorem
* $f(c)=\frac{F(a)-F(b)}{a-b}$

$f(c)(a-b)=F(a)-F(b)$

$\therefore f(c)(a-b)=\int_{a}^bf(x)\,dx$
