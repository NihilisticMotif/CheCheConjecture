#Squeeze theorem

Squeeze Theorem state that

If
* $\lim_{x\to a}l(x)=\lim_{x\to a}g(x)=L$
* $\forall x\in [x_1,x_2]\,:g_1(x)\leq f(x)\leq g_2(x)$

Implies that
* $\lim_{x\to a}f(x)=L$

**About the Image**

This image show a specific example of squeeze theorem in which
* Purple Line: $l(x)=\cos(x)$
* Blue Line: $f(x)=\frac{\sin(x)}{x}$
* Green Line: $g(x)=x^2+1$

It is true that
* $\forall x\in[-\pi,\pi]\,:l(x)\leq f(x)\leq g(x)$
* $\lim_{x\to 0}l(x)=\lim_{x\to 0}\cos(x)=1$
* $\lim_{x\to 0}g(x)=\lim_{x\to 0}x^2+1=1$

Implies that
* $\lim_{x\to 0}f(x)=\lim_{x\to 0}\frac{\sin(x)}{x}=1$

Note that
* $f(0)=\emptyset$

**About the Image**

There are 3 function
* Red Line: $f(x)=\frac{x}{\sin{x}}$
* Green Line: $l(x)=-\sin(x)+1$
* Blue Line: $g(x)=\sin(x)$

Even through the limit of every function when $x$ approach $0$ is equal to 1
* $\lim_{x\to0}f(x)=\lim_{x\to0}\frac{x}{\sin{x}}=1$
* $\lim_{x\to0}l(x)=\lim_{x\to0}-\sin(x)+1=1$
* $\lim_{x\to0}g(x)=\lim_{x\to0}\sin(x)=1$

But squeeze theorem cannot be applied to find $\lim_{x\to0}f(x)$ because
* $\exists x\in[x_1,x_2]\,:l(x)\geq f(x)\geq g(x)$
* $\exists x\in[x_1,x_2]\,:l(x)\leq f(x)\leq g(x)$

In addition the interval $[x_1,x_2]$ must include $a$ the other number near $a$ such that
* $a\in[x_1,x_2]$
* $a\neq x_1\land a\neq x_2$
