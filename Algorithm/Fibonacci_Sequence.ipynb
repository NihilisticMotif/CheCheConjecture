##Geometric Series

Find
* $\Sigma_{i=1}^np^i=\,?$

###Solution

Define
* $f(n)=\Sigma_{i=1}^np^i$

Note that
* $f(n+1)=f(n)+p^{n+1}$

Calculate
* $f(n+1)=\Sigma_{i=1}^{n+1}p^i$
* $f(n+1)=p(\Sigma_{i=0}^{n}p^i)$
* $f(n+1)=p(1+\Sigma_{i=1}^{n}p^i)$
* $f(n+1)=p(1+f(n))$

Calculate
* $f(n)+p^{n+1}=p(1+f(n))$
* $f(n)+p^{n+1}=p+pf(n)$
* $p^{n+1}-p=pf(n)-f(n)$
* $p^{n+1}-p=(p-1)\cdot f(n)$
* $\therefore f(n)=\frac{p\cdot(p^n-1)}{p-1}$

Therefore
* $\Sigma_{i=1}^np^i=\frac{p\cdot(p^n-1)}{p-1}$


###Geometric Series in Special Case

According to Geometric Series
* $\Sigma_{i=1}^np^i=\frac{p\cdot(p^n-1)}{p-1}$

Then
* $\Sigma_{i=1}^n2^i=2\cdot(2^n-1)$
* $\Sigma_{i=1}^n2^i=2^{n+1}-2$

Implies that
* $\forall S\subset\{1,...,n\}: \Sigma_{s\in S}2^s\leq 2^{n+1}-2 < 2^{n+1}$
* $\forall S\subset\{1,...,n\}: \Sigma_{s\in S}2^s < 2^{n+1}$

##Space Complexity

###Solution

Define
* $O(m)$ means Space Complexity of the Algorithm based in the recursive call stack.
* $\forall m\in\mathbb{N}:O(\,\bigcirc_{i=1}^mf(a_i)\,)=m$

Note that
* $\forall n\geq2:F(n)=F(n-1)+F(n-2)$
* $\forall n\geq2:O(\,F(n)\,)=O(\,F(n-1)\,)+1$

Note that
* $F(1)=F(0)=1$

Implies that the maximum possible size of recursive call stack of Fibonacci sequence with $n$ as the input, is $n$.

Therefore the space complexity of this algorithm is $O(n)$

#Method No.$2$ with $O(n)$ Time and Space



###Implementation

````python
def Fib(n,ls,Sigma):
  # Time : O(n)
  # Space: O(n)
  if n in [0,1]:
    return 1
  Fib(n-1,ls,Sigma)
  if len(ls) < 2:
    ls.append(n)
  else:
    Prev=ls[0]
    Temp=ls[1]
    ls[0]=Temp
    ls[1]+=Prev
  return ls[-1]

def Fibonacci(i):
  return Fib(i,[],0)

for i in range(11):
  print(i,Fibonacci(i))
````

This is because it takes at most $2$ times for appending and popping the recursive stack call, such that the program return the last element `ls[-1]` which is identical to the input `n`. The space and time complexity is also $O(1)$ in this case.

###General Case

It takes $n$ times and space to append the stack call in this line
* `Fib(n-1,ls,Sigma)`

When `n==1` the program return $1$

```
if n in [0,1]:
  return 1
```

The stack call pop the last element with Last In First Out methods which takes $n$ times. When the stack call is poped for the first $2$ times, `ls` is appended, such that

```
if len(ls) < 2:
  ls.append(n)
```

When `ls` have length equal $2$, `ls` is updated such that
* Updated `ls[0]` becomes Previous `ls[1]`
* Updated `ls[1]` becomes Previous (`ls[0]+ls[1]`)

```
Prev=ls[0]
Temp=ls[1]
ls[0]=Temp
ls[1]+=Prev
```

Because every recursive call return `1` for the based case, and return `ls[-1]` for the general case, the last recursive call return the valid output of Fibonacci Sequence regarding to any positive integer input `n` using $O(n)$ time and space.

#Method No.$3$ with $O(n)$ Time and $O(1)$ Space


##Implementation

````python
def Fib(n):
  # Time: O(n)
  # Space: O(1)
  Prev=1
  Sigma=2
  for i in range(n+1):
    if i>2:
      Temp=Sigma
      Sigma+=Prev
      Prev=Temp
  return Sigma

def Fibonacci(n):
  if n in [0,1]:
    return 1
  else:
    return Fib(n)

for i in range(11):
  print(i,Fibonacci(i))
````

#Method No.$4$ with $O(\log(n)^2)$ Time and $O(\log(n))$ Space

##Implementation

````python
def GetDigit(n):
  # Time : O(log(n))
  # Space: O(log(n))
  m=str(bin(n))[2:]
  Digit=[]
  for i in m:
    Digit.append(int(i))
  return Digit

def Fib(n):
  # Time : O(log(n)^2)
  # Space: O(log(n))
  Matrix=np.array([
    [1,1],
    [1,0]
  ])
  Product=np.identity(2)
  Position=1
  Digit=GetDigit(n)
  for digit in Digit:
    if digit==1:
      Partial=Matrix
      Index=2**(len(Digit)-Position)
      i=2
      while i<=Index:
        Partial=np.dot(Partial,Partial)
        i=2*i
      Product=np.dot(Product,Partial)
    Position+=1
  return Product[0][0]
````

````python
for i in range(20):
  print(i,Fib(i))
````

##Complexity of `GetDigit`

According to Geometric Series
* $\Sigma_{i=1}^m2^i=2\cdot(2^m-1)$

Given that
* $2^m < \Sigma_{i=1}^m2^i < 2^{m+1}$

Implies that it takes $m$ times and space to append binary digit of $n$. Therefore the time and space complexity of this algorithm is $O(\log_2(n))$

* $[\,2^m=n\iff \log_2(n)=m\,]\implies O(m)=O(\log_2(n))$

* $F(2)=2$

Define Fibonacci Number in General Case
* $\forall n\geq2:F(n)=F(n-1)+F(n-2)$

Calculate Based Case

* $\begin{bmatrix}
1&1\\
1&0\end{bmatrix}^2=\begin{bmatrix}
1&1\\
1&0\end{bmatrix}\cdot\begin{bmatrix}
1&1\\
1&0\end{bmatrix}=\begin{bmatrix}
2&1\\
1&1\end{bmatrix}$

Calculate General Case

* $\begin{bmatrix}
1&1\\
1&0\end{bmatrix}\cdot\begin{bmatrix}
F(n-1)&F(n-2)\\
F(n-2)&F(n-3)\end{bmatrix}=\begin{bmatrix}
F(n-1)+F(n-2)&F(n-2)+F(n-3)\\
F(n-1)&F(n-2)\end{bmatrix}$

* $\begin{bmatrix}
1&1\\
1&0\end{bmatrix}\cdot\begin{bmatrix}
F(n-1)&F(n-2)\\
F(n-2)&F(n-3)\end{bmatrix}=\begin{bmatrix}
F(n)&F(n-1)\\
F(n-1)&F(n-2)\end{bmatrix}$

* $\begin{bmatrix}
1&1\\
1&0\end{bmatrix}\cdot\begin{bmatrix}
F(n-1)&F(n-2)\\
F(n-2)&F(n-3)\end{bmatrix}=\begin{bmatrix}
1&1\\
1&0\end{bmatrix}^{n}$

Note that in some context, the author define $F(0)=0$. However I define $F(0)=1$ because
* $\begin{bmatrix}
1&1\\
1&0\end{bmatrix}^0=\begin{bmatrix}
1&0\\
0&1\end{bmatrix}$

###Base Case

Given that time and space complexity of `GetDigit(n)` is $O(\log_2(n))$ implies that it takes $O(\log_2(n))$ times to loop `Digit`
* `for digit in Digit:`

If $n=0$ because `np.identity(2)[0][0] = 1` $\implies $ `Fib(0)` $=1$

If $n=1$ because `Matrix[0][0] = 1` $\implies $ `Fib(0)` $=1$

###Other Case

####If $n=2^m$

Note that
* `Digit[0] == 1`
* `Index == 2**(m+1-1)`

Then, executing while loop in $m$ times
* `i = 2*i`
* `Partial^{2} == np.dot(Partial,Partial)`
* `Partial^{2*i} == ( Partial^{i} )^{2}`

Lastly update `Product` such that
* `Product = np.dot(Product,Partial)`
* $n=2^m\implies$ `Product = np.dot(np.identity(2),Partial)`

Return value is
* `Fib(2**m) == Product[0][0]`

####If $n=2^m+1$

The while loop is not activated because `i > Index`
* `Index=2**(0) == 1`
* `i=2`

The `Product` will be updated such that
* `Partial = Matrix`
* `Product = np.dot(Product,Partial)`

Return value is
* `Fib(2**m+1) == Product[0][0]`

####If $n=2^m+2$

The while loop is activated a single time, such that
* `Partial = Matrix`
* `Partial = np.dot(Partial,Partial)`
* `Product = np.dot(Product,Partial)`

Return value is
* `Fib(2**m+2) == Product[0][0]`

###General Case

The while loop is activated `len(Digit)-Position` times, such that
* `Partial = np.dot(Partial,Partial)`
* `i = 2*i`

After each while loop
* `Partial == Matrix^{len(Digit)-Position}`
* `Product = np.dot(Product,Partial)`
* `Position += 1`

Return value is
* `Fib(m) == Product[0][0]`

###Evaluation

Because the `Fib` contains $O(\log_2(n))$ times nested loops within $O(\log_2(n))$ times loop, the time complexity of `Fib` is $O(\log_2(n)^2)$ and the space complexity of `Fib` is $O(\log_2(n))$.





#Method No.$5$ with $O(\log(n))$ Time and Space

##Implementation

````python
def Fib(n):
  # Time : O(log(n))
  # Space: O(log(n))

  # Based Case
  Matrix=np.array([
    [1,1],
    [1,0]
  ])
  if n == 0:
    return np.identity(2)
  if n == 1:
    return Matrix

  # General Case
  Sigma = Fib(math.floor(n/2))
  Sigma = np.dot(Sigma,Sigma)
  if n%2==1:
    Sigma = np.dot(Sigma,Matrix)
  return Sigma

def Fibonacci(n):
  return Fib(n)[0][0]
````

````python
for i in range(20):
  print(i,Fibonacci(i))
````
