##**Set**

Set is the collection of elements.

* $a\in S$, this means $a$ is the element of set $A$
* $x\notin S$, this mean set $S$ does not contain $x$
* $2,3,5,7\in P_{10}$ is the set of prime number less than $10$ which is also finite set.
* $1,2,3,4,...\in\mathbb{N}$ is the set of all natural number, which is infinite set.


***
##**Important Set**
* $\emptyset=\{\}$
* $\mathbb{N}=\{1,2,3,...\}$
* $\mathbb{Z}=\{...,-3,-2,-1,01,2,3,...\}$
* $\mathbb{Q}=\{\frac{p}{q}|p\in\mathbb{Z},q\in\mathbb{N}\}$
* $\mathbb{R}$
* $\mathbb{C}=\{a+bi|a,b\in\mathbb{R},i^2=-1\}$


***
##**Cardinality**
* Number of all elements inside the set.
* $2,3,5,7\in P_{10}\implies |P_{10}|=4$
* $|\emptyset|=0$
* $|\emptyset,\{\emptyset\},\{\emptyset\{\emptyset\}\}|=3$
* $|\mathbb{R},\mathbb{C},\mathbb{N}|=3$


***

##**Example**
* $\{3n|n\in\mathbb{N}\}=\{3,6,9,12,...\}$
* $\{x\in\mathbb{R}|x^2=b\}=\{\sqrt{b},-\sqrt{b}\}$
* $\{x\in\mathbb{R}||x|\leq a\}=[-a,a]$

##**Binomial Coefficients**

##Formula
* $\binom{n}{k}=\frac{n!}{k!(n-k)!},0\leq k\leq n$

##Idea

###Given that
* $N=\{n_1,n_2,...,n_n\}$
 * $n=|N|$
* $S=(n_1,n_2,...,n_s),n_i\neq n_j$
 * $s=|S|\leq n$
* $\therefore $ number of all possible list of $S=n(n-1)(...)(n-s+1)=\frac{n(n-1)(...)(n-s+1)(n-s)(...)(1)}{(n-s)(n-s-1)(...)(1)}=\frac{n!}{(n-s)!}$

###Given that
* $N=\{n_1,n_2,...,n_n\}$
 * $n=|N|$
* If 2 sets have same set of elements, both sets are the same set. (The arrangement does not matter)
 * $K\subset N$
 * $k=|K|\leq n$
 * $K=\{n_1,n_2,...,n_k\}=\{n_2,n_1,...,n_k\}$
 * There are $k!$ possible arrangements of $K$
* $\binom{n}{k}=$ number of all possible set $K$ that $K\subset N$
 * $\therefore \binom{n}{k}=\frac{n!}{k!(n-k)!}$
 * Therefore there are $\frac{n!}{k!(n-k)!}$ possible subset with $k$ elements

##Example

* $N=\{1,2,3,4\}$
* $n=|N|=3$

$\begin{bmatrix}
k & \text{SubSet} & \binom{n}{k}\\
0 & \emptyset & 1\\
1 & \{1\},\{2\},\{3\},\{4\} & 4\\
2 & \{1,2\},\{1,3\},\{1,4\},\{2,3\},\{2,4\},\{3,4\} & 6\\
3 & \{2,3,4\},\{1,3,4\},\{1,2,4\},\{1,2,3\} & 4\\
4 & \{1,2,3,4\} & 1\\
k>4 & - & 0\\
\end{bmatrix}$

##**Power Set**

Power set of $A$ is the set of all subsets of $A$
* $\mathcal{P}(A)$

Example
* $A=\{0,1,2\}\implies \mathcal{P}(A)=\{\emptyset,\{0\},\{1\},\{2\},\{0,1\},\{0,2\},\{1,2\},\{0,1,2\}\}$

##**Size of Power Set**
* $|\mathcal{P}(A)|=2^{|A|}$








###Prove the size of the power set.

Define
* $|A|=n$

Given that
* $\binom{n}{k}=\frac{n!}{k!(n-k)!}$
* $|\mathcal{P}(A)|=\Sigma_{k=0}^{n}\binom{n}{k}$

Suppose that
* $\Sigma_{k=0}^{n}\binom{n}{k}=2^{n}$

If: $n=0\implies |\mathcal{P}(A)|=\binom{0}{0}=1$

If: $n=1\implies |\mathcal{P}(A)|=\binom{1}{0}+\binom{1}{1}=2=2^1$

If: $n=m+1\implies |\mathcal{P}(A)|=\binom{m+1}{0}+\binom{m+1}{1}+...+\binom{m+1}{m+1}$

$|\mathcal{P}(A)|=2+\binom{m+1}{1}+...+\binom{m+1}{m}$

$|\mathcal{P}(A)|=2+\Sigma_{k=1}^{m}\binom{m+1}{k}$

Given that
* $\binom{m+1}{k}=\binom{m}{k}+\binom{m}{k-1}$

Implies that
* $|\mathcal{P}(A)|=2+\Sigma_{k=1}^{m}\binom{m+1}{k}=2+\Sigma_{k=1}^m[\binom{m}{k}+\binom{m}{k-1}]$

$|\mathcal{P}(A)|=2+\Sigma_{k=1}^m\binom{m}{k}+\Sigma_{k=1}^m\binom{m}{k-1}$

$|\mathcal{P}(A)|=2+\Sigma_{k=1}^m\binom{m}{k}+\Sigma_{k=0}^{m-1}\binom{m}{k}$

$|\mathcal{P}(A)|=2+\Sigma_{k=1}^{m-1}\binom{m}{k}+\binom{m}{m}+\Sigma_{k=1}^{m-1}\binom{m}{k}+\binom{m}{0}$

$|\mathcal{P}(A)|=4+\Sigma_{k=1}^{m-1}\binom{m}{k}+\Sigma_{k=1}^{m-1}\binom{m}{k}$

$|\mathcal{P}(A)|=4+2*\Sigma_{k=1}^{m-1}\binom{m}{k}$

$|\mathcal{P}(A)|=2(2+\Sigma_{k=1}^{m-1}\binom{m}{k})$

$|\mathcal{P}(A)|=2(\binom{m}{0}+\binom{m}{m}+\Sigma_{k=1}^{m-1}\binom{m}{k})$

$|\mathcal{P}(A)|=2(\Sigma_{k=0}^{m}\binom{m}{k})$

Given that
* $\Sigma_{k=0}^{n}\binom{n}{k}=2^{n}$

$|\mathcal{P}(A)|=2(2^m)$

$|\mathcal{P}(A)|=2^{m+1}$

Given that
* $|A|=n=m+1$

Therefore
* $|\mathcal{P}(A)|=2^{|A|}$

##Exercise No.4

Given that
* $b< a< c$
* $x\in\mathbb{Z}$
* $A=\{ax^2+bx< c |x\in\mathbb{Z}\}$
* $|A|=?$


###Solution

According to quadratic formula
* $ax^2+bx+c=0\implies x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}$

Given that
* $ax^2+bx-c=0\implies x=\frac{-b\pm\sqrt{b^2+4ac}}{2a}$
* $A=\{ax^2+bx< c |x\in\mathbb{R}\}\cap\mathbb{Z}$

Therefore
* $k=\frac{-b+\sqrt{b^2+4ac}}{2a}-\frac{-b-\sqrt{b^2+4ac}}{2a}$
* $k=\frac{-b+\sqrt{b^2+4ac}}{2a}+\frac{b+\sqrt{b^2+4ac}}{2a}$
* $k=\frac{\sqrt{b^2+4ac}}{2a}+\frac{\sqrt{b^2+4ac}}{2a}$
* $k=\frac{2\sqrt{b^2+4ac}}{2a}$
* $k=\frac{\sqrt{b^2+4ac}}{a}$

Therefore
* $|A|=⌊ k ⌋-1$
